```python
import requests
import csv
import os

list1 = []
with open('link.csv', encoding = 'utf-8') as file1:
    for row in csv.reader(file1):
        list1.append(row[0])
file1.close()

# Loop through the list of links and make requests
for i in range(1, len(list1)):
    url = "https://www.bilibili.com/video/" + list1[i]
    
    # Set appropriate headers
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    try:
        res = requests.get(url, headers=headers)
        res.raise_for_status()  # Raise an error for bad responses (4xx and 5xx)
        
        res.encoding = 'utf-8'
        html = res.text
        filename = str(i) + "_" + list1[i] + ".txt"  # Adjusted filename for safety
        filepath = "./bili file/" + filename
        
        with open(filepath, "w", encoding='utf-8') as file2:
            file2.write(html)

    except requests.exceptions.HTTPError as err:
        print(f"HTTP error occurred for URL {url}: {err}")
    except Exception as e:
        print(f"An error occurred for URL {url}: {e}")
```
